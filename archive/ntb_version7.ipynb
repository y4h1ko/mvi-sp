Extensive tuning of the flow does not significantly reduce the gap between predicted mean and true frequencies for N=1000.
Increasing the dataset size improves accuraccy, but does not move mean closer to true frequency, but signal length, noise level, and discretization could help (not sure about reducing gap between mean and true value). Maybe combining likelihood with additional constraints could help, but I am not sure for purely likelihood-based probabilistic model.
