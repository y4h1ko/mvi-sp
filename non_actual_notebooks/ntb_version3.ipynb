<span style="font-size: 30px;">Playing with different N and t_disc</span>

- I did iteration through several total samples (from 1000 to 10000) and with different discretization of time (form 60 to 150 levels) with previous sufficient EPOCHS = 160

```python
#with this i tried to find how N and t_disc affect loss function
N_values = [1000, 2500, 5000, 7500, 10000]
t_disc_values = [60, 70, 85, 100, 125, 150]
```

This will iterate through all variations of values and then show plots (full size and zooms)

```python
MAX_EPOCHS_FOR_SEARCH = 160

results = sweep_N_and_tdisc(N_values, t_disc_values, max_epochs=MAX_EPOCHS_FOR_SEARCH)
print_results_table(results)

# plots
for N_val in N_values:
    plot_val_curves_fixed_N(results, N=N_val)
    plot_val_curves_fixed_N(results, N=N_val, y_limit=0.05, zoom=f"0.05")
    plot_val_curves_fixed_N(results, N=N_val, y_limit=0.01, zoom=f"0.01")
```
